#support vector machine

support vector machine draws a hyper plane in n dimensional space such that it maximize the margine between the clissification groups 

Gamma: Gamma is a hyperparameter which we have to set before training model. Gamma decides that how much curvature we want in a decision boundary.
Gamma high means more curvature.
Gamma low means less curvature.

Regularization: The Regularization parameter (often termed as C parameter in python’s sklearn library) tells the SVM optimization how much you want to avoid misclassifying each training example.

Kernal: Kernel Function generally transforms the training set of data so that a non-linear decision surface is able to transformed to a linear equation in a higher number of dimension spaces.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
#building the DataFrame frame
df = pd.DataFrame(irish.data, columns=irish.feature_names)
df.head()
```

```python
#appending the target 
df['target'] = irish.target
df
```

```python
irish.target_names
```

0 indicate setosa 1 indicates versicolor 2 indicates virginica

```python
df['flower_name'] = df.target.apply(lambda x:irish.target_names[x])
df
```

Data visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'red')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker='+')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'red')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'blue', marker='+')
```

```python
X = df.drop(['target', 'flower_name'], axis = 1)
Y = df['target']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.target
```

```python
dfd = pd.DataFrame(digits.data, digits.target)
dfd.head()
```

```python
dfd['target'] = digits.target
dfd
```

```python
X_trainE, X_testE, y_trainE, y_testE = train_test_split(dfd.drop('target',axis='columns'), dfd.target, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
modelE = SVC()
```

```python
modelE.fit(X_trainE, y_trainE)
```

```python
modelE.score(X_testE, y_testE)
```

```python
model_L = SVC(kernel='linear')
```

```python
model_L.fit(X_trainE, y_trainE)
```

```python
model_L.score(X_testE, y_testE)
```
