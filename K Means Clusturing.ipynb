```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv")
data.head()
```

```python
plt.scatter(data['Age'], data['Income($)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
Y_predicted = km.fit_predict(data[['Age', 'Income($)']])
Y_predicted
```

```python
data['cluster'] = Y_predicted
data.head()
```

```python
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]

plt.scatter(data1.Age, data1['Income($)'], color='green')
plt.scatter(data2.Age, data2['Income($)'], color='red')
plt.scatter(data3.Age, data3['Income($)'], color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
scaler.fit(data[['Income($)']])
data['Income($)'] = scaler.transform(data[['Income($)']])

scaler.fit(data[['Age']])
data['Age'] = scaler.transform(data[['Age']])

data
```

```python
km = KMeans(n_clusters=3)
Y_predicted = km.fit_predict(data[['Age', 'Income($)']])
Y_predicted
```

```python
km.cluster_centers_   #gives the centroids of the cluster
```

```python
data['cluster'] = Y_predicted
data
```

```python
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]

plt.scatter(data1.Age, data1['Income($)'], color='green')
plt.scatter(data2.Age, data2['Income($)'], color='red')
plt.scatter(data3.Age, data3['Income($)'], color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'purple', marker='*', label = 'centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df.drop(df.iloc[:,2:4], axis=1, inplace=True)
df
```

```python
#making the clusters

from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
```

```python
y_predict = km.fit_predict(df)
y_predict
```

```python
df['cluster'] = y_predict
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = df[df['cluster'] == 0]
df2 = df[df['cluster'] == 1]
df3 = df[df['cluster'] == 2]

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'red')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'green')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'], color = 'black')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```

The scatter plot shows us that still the clusters are not forms perfectly so we need to scale the values to make the cluster perfect for that we will use minmax scaller

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(df[['sepal length (cm)']])
df['sepal length (cm)'] = scaler.transform(df[['sepal length (cm)']])

scaler.fit(df[['sepal width (cm)']])
df['sepal width (cm)'] = scaler.transform(df[['sepal width (cm)']])

df
```

```python
y_predict = km.fit_predict(df)
y_predict
```

```python
df1 = df[df['cluster'] == 0]
df2 = df[df['cluster'] == 1]
df3 = df[df['cluster'] == 2]

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'red')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'green')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'], color = 'black')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```
