```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv')
data.head()
```

```python
data.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True)
data.head()
```

```python
targets = data.Survived
inputs = data.drop('Survived', axis = 1)
```

```python
dummies = pd.get_dummies(inputs['Sex'])
dummies.head()
```

```python
inputs = pd.concat([inputs, dummies], axis = 'columns')
inputs.head()
```

```python
inputs.drop('Sex', axis = 1, inplace = True)
inputs.head()
```

```python
inputs.isnull().sum()
```

```python
inputs['Age'] = inputs['Age'].fillna(inputs.Age.mean())
```

```python
inputs.isnull().sum()
```

```python
targets.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size = 0.2, random_state = 42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X_test[0:10]
```

```python
y_test[0:10]
```

```python
model.predict(X_test[0:10])
```

```python
model.predict_proba(X_test[0:10])
```

Spam checker

```python
spam = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv')
spam.head()
```

```python
spam.groupby('Category').describe()
```

```python
spam['spam'] = spam['Category'].apply(lambda x: 1 if x=='spam' else 0)
spam.head()
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(spam.Message, spam.spam, test_size = 0.2, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
                ('vectorizer', CountVectorizer()),
                ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.predict(emails)
```
